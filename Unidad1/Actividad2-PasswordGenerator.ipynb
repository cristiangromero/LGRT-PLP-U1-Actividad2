```python
class PasswordTransformer:
    """
    Clase base. Extrae las iniciales de cada palabra
    y alterna mayúsculas y minúsculas.
    """

    def __init__(self, phrase: str):
        self._phrase = phrase       # frase ingresada por el usuario
        self._password = ""         # contraseña generada (se completa al transformar)

    def _extract_initials(self) -> str:
        """Devuelve la primera letra de cada palabra unidas en un string."""
        return "".join(word[0] for word in self._phrase.split() if word)

    def _alternate_case(self, text: str) -> str:
        """Alterna mayúscula (par) y minúscula (impar) según la posición del caracter."""
        return "".join(c.upper() if i % 2 == 0 else c.lower() for i, c in enumerate(text))

    def transform(self) -> str:
        """Aplica las transformaciones base (primeras letras + may. min. según posición) y guarda el resultado."""
        self._password = self._alternate_case(self._extract_initials())
        return self._password

    def __str__(self) -> str:
        return (f"  Frase original    : {self._phrase}\n"
                f"  Contraseña segura : {self._password}")


class SecurePasswordTransformer(PasswordTransformer):
    """
    Clase hija de PasswordTransformer agregando
    sustitución de letras por caracteres especiales.
    """

    # Tabla de sustituciones: letra -> caracter especial
    SUBSTITUTIONS = {'a': '@', 'e': '3', 'i': '!', 'o': '0', 's': '$', 't': '+'}

    def __init__(self, phrase: str):
        super().__init__(phrase)    # inicializa los atributos heredados

    def _substitute_chars(self, text: str) -> str:
        """Reemplaza letras por caracteres especiales según la tabla SUBSTITUTIONS."""
        return "".join(self.SUBSTITUTIONS.get(c.lower(), c) for c in text)

    def transform(self) -> str:
        """Extiende la transformación base sumando la sustitución de caracteres."""
        base_password = super().transform()             # reutiliza la lógica del padre
        self._password = self._substitute_chars(base_password)
        return self._password


# -- Programa principal --

if __name__ == "__main__":

    phrase = input("Ingresá tu frase a transformar: ")

    transformer = SecurePasswordTransformer(phrase)
    transformer.transform()

    print("\n--- Resultado final ---")
    print(transformer)
```
